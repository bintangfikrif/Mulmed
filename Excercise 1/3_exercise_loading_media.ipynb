# Pendahuluan

## Apa itu Data Multimedia?

Data multimedia adalah informasi yang dikodekan dalam berbagai format untuk merepresentasikan dunia nyata:

- **Audio (1D)**: Sinyal satu dimensi yang berubah terhadap waktu
  - Contoh: musik, suara, speech
  - Representasi: amplitudo vs waktu
  
- **Gambar (2D)**: Matriks nilai intensitas dalam ruang dua dimensi
  - Contoh: foto, ilustrasi, grafik
  - Representasi: intensitas pixel pada koordinat (x,y)
  
- **Video (2D + Waktu)**: Rangkaian frame (gambar) yang ditampilkan berurutan
  - Contoh: film, rekaman, animasi
  - Representasi: frame berubah terhadap waktu dengan frame rate tertentu

### A2. TODO: Muat & Metadata

**Instruksi:**
Tulis kode untuk memuat file audio dan menampilkan metadata dasar:
- Sample rate (Hz)
- Durasi (detik)
- Jumlah kanal (mono/stereo)
- Jumlah total sampel

---



```python
# Muat file audio dan tampilkan metadata
AUDIO_PATH = os.path.join(os.getcwd(), 'data', 'whale_sound.wav')

y = None
sr = None
source_info = ""

try:
    if os.path.exists(AUDIO_PATH):
        y, sr = librosa.load(AUDIO_PATH, sr=None)
        source_info = f" Berhasil Memuat {AUDIO_PATH}"
    else:
        source_info = f" File tidak ditemukan"
except Exception as e:
    print(f"Error loading audio: {e}")

# Konversi ke mono jika stereo
if y.ndim > 1:
    y = np.mean(y, axis=0)
    source_info += " (dikonversi ke mono)"

y = y.astype(np.float32)

print(source_info)
print(f"Shape data: {y.shape}")
print(f"Sample rate: {sr} Hz")
print(f"Durasi: {len(y)/sr:.2f} detik")
print(f"Jumlah kanal: {'Mono (1 kanal)' if y.ndim == 1 else f'Multi-kanal ({y.ndim} kanal)'}")
print(f"Tipe data: {y.dtype}")
```

### A3. TODO: Waveform

**Instruksi:**
Plot waveform audio dengan:
- Sumbu X: waktu (detik)
- Sumbu Y: amplitudo
- Judul dan label sumbu yang jelas

```python
# Plot Waveform dengan Detail Zoom
# Buat vektor waktu
t = np.linspace(0, len(y)/sr, len(y))

# Setup plot dengan 2 subplot
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 8))

# Plot 1: Waveform lengkap
ax1.plot(t, y, color='blue', linewidth=0.5)
ax1.set_title('Waveform Audio Lengkap', fontsize=14, fontweight='bold')
ax1.set_xlabel('Waktu (detik)')
ax1.set_ylabel('Amplitudo')
ax1.grid(True, alpha=0.3)
ax1.set_ylim([-1.1, 1.1])

# Tambahkan informasi durasi
duration_text = f'Durasi total: {len(y)/sr:.2f} detik'
ax1.text(0.02, 0.95, duration_text, transform=ax1.transAxes,
         bbox=dict(boxstyle="round,pad=0.3", facecolor="lightblue", alpha=0.7))

# Plot 2: Zoom detail 0-0.05 detik (atau maksimal 10% dari durasi total)
zoom_duration = min(0.05, (len(y)/sr) * 0.1)
zoom_samples = int(zoom_duration * sr)
t_zoom = t[:zoom_samples]
y_zoom = y[:zoom_samples]

ax2.plot(t_zoom, y_zoom, color='red', linewidth=1.5, marker='o', markersize=2)
ax2.set_title(f'Detail Waveform (0 - {zoom_duration:.3f} detik)', fontsize=14, fontweight='bold')
ax2.set_xlabel('Waktu (detik)')
ax2.set_ylabel('Amplitudo')
ax2.grid(True, alpha=0.3)
ax2.set_ylim([-1.1, 1.1])

# Tambahkan informasi zoom
zoom_text = f'Menampilkan {zoom_samples} sampel pertama'
ax2.text(0.02, 0.95, zoom_text, transform=ax2.transAxes,
         bbox=dict(boxstyle="round,pad=0.3", facecolor="lightcoral", alpha=0.7))

plt.tight_layout()
plt.show()

print(f"Waveform berhasil ditampilkan!")
print(f"Total sampel yang diplot: {len(y):,}")
print(f"Sampel zoom detail: {zoom_samples}")
```

**Analisis yang diperlukan:**
Jelaskan apa yang Anda lihat dari waveform (pola amplitudo, bagian keras/pelan, dll.)

Gelombang audio di atas menggambarkan suara dari seekor paus, sering disebut juga sebagai "nyanyian paus".

Terdapat gelombang beramplitudo tinggi pada durasi 0-5 detik dan 24-30 detik yang mengartikan pada saat "nyanyian paus" ini sedang berlangsung.

Sedangkan pada detik-detik diantaranya terdapat gelombang beramplitudo rendah, artinya ini adalah jeda atau transisi antar bagian dari "nyanyian paus" itu.

### A4. TODO: Spectrogram log-dB

**Instruksi:**
Hitung STFT dan tampilkan spectrogram dalam skala log-dB:
- Gunakan parameter standar (n_fft=1024, hop_length=256)
- Tampilkan dengan colorbar
- Label sumbu: waktu (detik) dan frekuensi (Hz)


```python
# Parameter STFT
n_fft = 1024
hop_length = 256
window = 'hann'

# Hitung STFT
D = librosa.stft(y, n_fft=n_fft, hop_length=hop_length, window=window)
magnitude = np.abs(D)
magnitude_db = librosa.amplitude_to_db(magnitude, ref=np.max)

# Setup plot dengan 2 subplot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Plot 1: Magnitude Spectrogram (Linear)
img1 = librosa.display.specshow(magnitude, x_axis='time', y_axis='hz',
                               sr=sr, hop_length=hop_length, ax=ax1)
ax1.set_title('📊 Spectrogram - Magnitude Linear', fontsize=14, fontweight='bold')
ax1.set_xlabel('Waktu (detik)')
ax1.set_ylabel('Frekuensi (Hz)')
plt.colorbar(img1, ax=ax1, format='%+2.0f')

# Plot 2: Magnitude Spectrogram (Log dB)
img2 = librosa.display.specshow(magnitude_db, x_axis='time', y_axis='hz',
                               sr=sr, hop_length=hop_length, ax=ax2)
ax2.set_title('📊 Spectrogram - Magnitude (dB)', fontsize=14, fontweight='bold')
ax2.set_xlabel('Waktu (detik)')
ax2.set_ylabel('Frekuensi (Hz)')
plt.colorbar(img2, ax=ax2, format='%+2.0f dB')

plt.tight_layout()
plt.show()

# Informasi STFT
print("📋 INFORMASI STFT")
print("=" * 40)
print(f"🔢 Ukuran FFT: {n_fft}")
print(f"👣 Hop length: {hop_length}")
print(f"🪟 Window function: {window}")
print(f"📊 Shape magnitude: {magnitude.shape}")
print(f"⏱️ Resolusi waktu: {hop_length/sr*1000:.1f} ms per frame")
print(f"🎵 Resolusi frekuensi: {sr/n_fft:.1f} Hz per bin")
print(f"📈 Range magnitude (dB): {magnitude_db.min():.1f} - {magnitude_db.max():.1f}")
```

**Analisis yang diperlukan:**
Jelaskan perbedaan informasi yang didapat dari spectrogram dibanding waveform.

Secara umum Waveform dan Spectogram memberikan informasi yang berbeda, dengan rincian sebagai berikut.
- Waveform: Menunjukkan volume (amplitudo) dari suara paus dari waktu ke waktu.
- Spektogram: Memberikan detail lebih lanjut dengan menampilkan frekuensi (nada) dan intensitas suara seiring waktu.

Hal ini berarti kedua grafik di atas memiliki fungsi yang berbeda. Waveform efektif untuk mengidentifikasi keberadaan suara dan jeda antar suara, sedangkan spectogram penting untuk menganalisis komposisi dan karakteristik frekuensi dari "nyanyian" paus itu.

### A5. TODO: MFCC

**Instruksi:**
Hitung dan tampilkan minimal 13 koefisien MFCC sebagai heatmap:
- Sumbu X: waktu (frame)
- Sumbu Y: koefisien MFCC (1-13)
- Gunakan colorbar dan judul yang jelas

```python
# Hitung Mel-Spectrogram dan MFCC
# Parameter untuk Mel-spectrogram
n_mels = 64  # Jumlah mel bins
fmax = sr // 2  # Frekuensi maksimum (Nyquist frequency)

# Hitung Mel-spectrogram
mel_spec = librosa.feature.melspectrogram(y=y, sr=sr, n_fft=n_fft,
                                         hop_length=hop_length, n_mels=n_mels, fmax=fmax)
mel_spec_db = librosa.power_to_db(mel_spec, ref=np.max)

# Hitung MFCC (13 koefisien)
n_mfcc = 13
mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc, n_fft=n_fft,
                           hop_length=hop_length, n_mels=n_mels)

# Setup plot dengan 2 subplot
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10))

# Plot 1: Mel-Spectrogram
img1 = librosa.display.specshow(mel_spec_db, x_axis='time', y_axis='mel',
                               sr=sr, hop_length=hop_length, fmax=fmax, ax=ax1)
ax1.set_title('🎶 Mel-Spectrogram (dB)', fontsize=14, fontweight='bold')
ax1.set_xlabel('Waktu (detik)')
ax1.set_ylabel('Frekuensi Mel')
plt.colorbar(img1, ax=ax1, format='%+2.0f dB')

# Plot 2: MFCC (Heatmap)
img2 = librosa.display.specshow(mfcc, x_axis='time', ax=ax2)
ax2.set_title('🎯 MFCC (13 Koefisien)', fontsize=14, fontweight='bold')
ax2.set_xlabel('Waktu (detik)')
ax2.set_ylabel('Koefisien MFCC')
plt.colorbar(img2, ax=ax2)

plt.tight_layout()
plt.show()

# Informasi Mel & MFCC
print("📋 INFORMASI MEL-SPECTROGRAM & MFCC")
print("=" * 50)
print(f"🎶 Jumlah Mel bins: {n_mels}")
print(f"🎯 Jumlah MFCC: {n_mfcc}")
print(f"📊 Shape Mel-spectrogram: {mel_spec.shape}")
print(f"📊 Shape MFCC: {mfcc.shape}")
print(f"📈 Range Mel-spectrogram (dB): {mel_spec_db.min():.1f} - {mel_spec_db.max():.1f}")
print(f"📈 Range MFCC: {mfcc.min():.2f} - {mfcc.max():.2f}")
print()
print("✅ Semua fitur spektral berhasil dihitung!")
print("🎵 Mel-spectrogram: Representasi frekuensi sesuai persepsi manusia")
print("🎯 MFCC: Ringkasan spektral untuk machine learning")
```

**Analisis yang diperlukan:**
Interpretasi sederhana: apakah pola MFCC stabil atau berubah-ubah? Apa potensi maknanya?

Pola MFCC seperti terlihat pada grafik di atas cenderung berubah-ubah secara signifikan dari waktu ke waktu.

Makna dari pola yang berubah-ubah ini bisa berarti bahwa paus tidak hanya mengubah volume atau nada suaranya, melainkan "gaya" dari suaranya. Selain itu, pola yang berubah-ubah ini membuktikan bahwa "nyanyian" paus ini memiliki struktur yang rumit dan bukan sekedar suara tunggal yang berulang.

```python
# Load Video
PATH_VIDEO = os.path.join(os.getcwd(), 'data', 'nature.mp4')

# Muat video dengan OpenCV
cap = cv2.VideoCapture(PATH_VIDEO)
if not cap.isOpened():
    print(f"Error opening video file: {PATH_VIDEO}")
else:
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    fps = cap.get(cv2.CAP_PROP_FPS)
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    duration = frame_count / fps if fps > 0 else 0

    print(f"Berhasil Membuka Video Pemandangan Alam")
    print(f"Frame Rate     : {fps:.2f} FPS")
    print(f"Durasi         : {duration:.2f} detik")
    print(f"Resolusi       : {width} x {height} piksel")
    print(f"Klasifikasi Resolusi : {'SD' if width < 1280 else 'HD' if width < 1920 else 'Full HD' if width < 3840 else '4K' if width < 7680 else '8K'}")
    print(f"Total Frame    : {frame_count}")

    cap.release()
```

**Analisis yang diperlukan:**
Jelaskan pentingnya parameter-parameter tersebut untuk analisis video atau aplikasi tertentu.

Setiap parameter dalam analisis video memainkan peran kunci yang saling melengkapi. Berikut rinciannya:

- Resolusi: Menentukan kualitas gambar dan detail.

- Frame rate: Menentukan kelancaran gerakan.

- Jumlah total frame: Ukuran konten mentah video.

- Durasi: Menentukan panjang total video.

- Klasifikasi resolusi: Standar label kualitas yang mudah dikenali (HD, 4K, dll.).

### C3. TODO: Tampilkan 3 Frame (Awal–Tengah–Akhir)

**Instruksi:**
Ambil dan tampilkan 3 frame representatif:
- Frame pertama (index 0)
- Frame tengah (index ~total_frame/2)
- Frame terakhir (index total_frame-1)
- **Konversi BGR→RGB** sebelum ditampilkan
- Subplot dengan judul frame dan timestamp

```python
# Ambil dan tampilkan 3 frame representatif dari video (awal, tengah, akhir)
cap = cv2.VideoCapture(PATH_VIDEO)
if not cap.isOpened():
    print(f"Error opening video file: {PATH_VIDEO}")
else:
    idxs = [0, frame_count // 2, frame_count - 1]
    frames = []
    timestamps = []
    for idx in idxs:
        cap.set(cv2.CAP_PROP_POS_FRAMES, idx)
        ret, frame = cap.read()
        if ret:
            frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            frames.append(frame_rgb)
            timestamps.append(idx / fps)
        else:
            frames.append(None)
            timestamps.append(None)
    cap.release()

    plt.figure(figsize=(18, 6))
    titles = ['Frame Awal', 'Frame Tengah', 'Frame Akhir']
    for i, (img, ts, title) in enumerate(zip(frames, timestamps, titles)):
        plt.subplot(1, 3, i+1)
        if img is not None:
            plt.imshow(img)
            plt.title(f"{title}\nIndex: {idxs[i]}, Time: {ts:.2f} detik")
            plt.axis('off')
        else:
            plt.text(0.5, 0.5, 'Frame tidak tersedia', ha='center', va='center')
            plt.axis('off')
    plt.suptitle('Tiga Frame Representatif dari Video')
    plt.tight_layout(rect=[0, 0, 1, 0.93])
    plt.show()
```

**Analisis yang diperlukan:**
Deskripsikan perbedaan visual antar frame dan apa yang dapat dipelajari dari sampel frame ini.

Terdapat 3 frame yang mewakili video secara keseluruhan. Berikut rinciannya:

- Frame Awal (index 0): Menampilkan pemandangan alam pengunungan dan sungai dengan latar langit sore hari menjelang matahari terbenam.
- Frame Tengah (index 360): Menampilkan pemandangan sungai di tengah hutan dengan banyak bebatuan disekitarnya.
- Frame Akhir (index 720): Menampilkan pemandangan air terjun di tengah hutan.
